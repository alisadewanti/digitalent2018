```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
# Prepare a list of offices sorted by country, state, city.

run("SELECT country, state, city, territory, officeCode, phone, addressLine1, addressLine2, postalCode FROM offices order by country, state, city")
```

```python
# How many employees are there in the company?
run("SELECT COUNT(employeeNumber) FROM EMPLOYEES")
```

```python
# What is the total of payments received?
run("select sum(amount) from payments")
```

```python
# List the product lines that contain 'Cars'
run("select * from productlines where productLine like '%cars%'")
```

```python
#Report total payments for October 28, 2004.
# run('select * from payments where paymentDate<="2004-10-28" ORDER BY PAYMENTDATE desc')
run('select SUM(amount) from payments where paymentDate <= "2004-10-28"')
```

```python
#Report those payments greater than 'USD100,000' -> 'those payments' refer to previous report statement
run('select * from payments where amount>100000 and paymentDate<="2004-10-28" ORDER BY PAYMENTDATE desc')
```

```python
# List the products in each product line
run('select productLine,productName from products order by productLine')
```

```python
#How many products in each product line?
run('select productLine, count(productcode) as ProductNumber from products group by productLine')
```

```python
#What is the minimum payment received?
run('select min(amount) as MinPayment from payments')
```

```python
#List all payments greater than twice the average payment
#run('select 2*avg(amount) from payments')
run('select * from payments where amount > 2*(select avg(amount) from payments) order by amount')
```

```python
# What is the average percentage markup of the MSRP on buyPrice?

run('select * from products')
run('select avg((msrp-buyprice)/buyprice) * 100 as AvgMarkUp from products')
```

```python
# How many distinct products does ClassicModels sell?
run('select count(distinct productName) as DistinctProductNumber from products')
```

```python
# Report the name and city of customers who don't have sales representatives

run('select customerName, city, salesRepEmployeeNumber from customers where salesRepEmployeeNumber is null')
```

```python
# What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.
run("SELECT CONCAT(firstName,' ',lastName) AS fullName,  jobTitle FROM employees where jobTitle like'%VP%' OR jobTitle like '%Manager%'")
```

```python
# Which orders have a value greater than USD5,000?

#run("select * from orders join orderdetails on orders.orderNumber = orderdetails.orderNumber")
run("""
select orderNumber, sum(quantityOrdered*priceEach) as orderValue
from orderdetails 
group by orderNumber
having orderValue > 5000
order by orderValue
""")
```
