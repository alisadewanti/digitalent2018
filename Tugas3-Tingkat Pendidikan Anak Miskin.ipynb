## 1. Deskripsi Data 

Dataset ini berisi data mengenai jumlah anak dengan status kesejahteraan 30% yang bersekolah berdasarkan tingkat pendidikan dan jenis kelamin. Data disajikan pada bulan Juli 2012.

Penjelasan mengenai Variabel pada Dataset ini:
- kode_provinsi: Kode Provinsi
- nama_provinsi: Nama Provinsi
- tingkat_pendidikan: Tingkat Pendidikan (SD/MI/SMP/MT/SMA/MA/Perguruan Tinggi)
- jenis_kelamin: Jenis Kelamin
- jumlah_individu: Jumlah dari Anak dengan Tingkat Pendidikan dan Jenis Kelamin tersebut

```python
import pandas as pd

data_pendidikan = pd.read_csv('processed-jumlah-partisipasi-sekolah-jenis-kelamin-30persen-termiskin-indonesia-perjuli2012.csv', delimiter = ',', encoding = "utf-8")
data_pendidikan
```

## 3. Visualisasi sederhana

```python
ax = data_pendidikan.groupby('nama_provinsi').jumlah_individu.sum()

ax = ax.plot(kind='bar', title ="Jumlah Anak Miskin Berdasarkan Provinsi", fontsize=12, figsize=(20,10))
ax.set_xlabel("Nama Provinsi", fontsize=12)
ax.set_ylabel("Jumlah Individu", fontsize=12)
```

```python
ax = data_pendidikan.groupby('tingkat_pendidikan').jumlah_individu.sum()

ax = ax.plot(kind='bar', title ="Jumlah Anak Miskin Berdasarkan Tingkat Pendidikan", fontsize=12)
ax.set_xlabel("Tingkat Pendidikan", fontsize=12)
ax.set_ylabel("Jumlah Individu", fontsize=12)
```

```python
ax = data_pendidikan.groupby('jenis_kelamin').jumlah_individu.sum()

ax = ax.plot(kind='pie', title ="Jumlah Anak Miskin Berdasarkan Jenis Kelamin", legend=True, fontsize=12)
ax.set_xlabel("Jenis Kelamin", fontsize=12)
ax.set_ylabel("Jumlah Individu", fontsize=12)
```

```python
import matplotlib.pyplot as plt

data=data_pendidikan.loc[(data_pendidikan['nama_provinsi']!='0')].pivot_table('jumlah_individu','nama_provinsi', 'jenis_kelamin', aggfunc=sum)
data.plot.bar(figsize=(20,10), title ="Perbandingan Jumlah Anak Miskin Pada Tiap Provinsi Berdasarkan Jenis Kelamin")
```

```python
import numpy as np

data=data_pendidikan.loc[(data_pendidikan['jenis_kelamin']!='0')].pivot_table('jumlah_individu','jenis_kelamin','tingkat_pendidikan', aggfunc=sum)

plt.figure()
data.plot(title="Perbandingan Jumlah Anak Miskin Berdasarkan Tingkat Pendidikan dan Jenis Kelamin", figsize=(10,5))
plt.xticks(np.arange(len(data.index)), data.index)
```

```python
import numpy as np
data=data_pendidikan.loc[(data_pendidikan['nama_provinsi']!=0)].pivot_table('jumlah_individu','nama_provinsi','tingkat_pendidikan', aggfunc=sum)
#data.plot.bar(figsize=(20,10))
plt.figure()
data.plot(figsize=(20,10),legend=True, rot=90, marker='o')
plt.xticks(np.arange(len(data.index)), data.index)
```
